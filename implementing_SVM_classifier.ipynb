```python
# bias value is -ve because b is towards yhe -ve axis
class svm_classifier():

    def __init__(self, learning_rate, no_of_iterations, lambda_parameter):#initiating hyper parameteer
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations
        self.lambda_parameter = lambda_parameter

    def fit(self, X, Y):#fitting the dataset to d=svm clssiier
        self.m, self.n = X.shape#m=no. of datpoints-no. of rows, #n=no. of input features- no, of rows
        self.w = np.zeros(self.n)#intiating weingt value which is array
        self.b = 0 #intiating bias value which is single value
        self.X = X
        self.Y = Y

        #imnplementing gradient descent algoruthm for optimization
        for i in range(self.no_of_iterations):
            self.update_weights()

    def update_weights(self):#funcyion to update weighnt and biaqs value
        y_label = np.where(self.Y <= 0, -1, 1)#SV tries to find label value is -1 or 1

        for index, x_i in enumerate(self.X):#[y=mx+c]
            condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

            if(condition == True):
                dw = 2 * self.lambda_parameter * self.w
                db = 0
            else:
                dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])

            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db

    def predict(self, X):#predict the label for a given input value
        output = np.dot(X, self.w) - self.b
        predicted_labels = np.sign(output)# sign is used either output is +1 or-1

        y_hat = np.where(predicted_labels <= -1, 0, 1)

        return y_hat
```
