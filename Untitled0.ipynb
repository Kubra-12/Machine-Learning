plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy', scale=1)
plt.quiver(0,0,-2,-5, scale_units='xy',angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector1=np.array([2,2])
vector2=np.array([3,-2])
add= vector1+vector2
plt.quiver(0,0,2,2, scale_units='xy',angles='xy', scale=1)
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy', scale=1)
plt.quiver(0,0,5,1, scale_units='xy',angles='xy', scale=1)
plt.xticks()  # Show default ticks on x-axis
plt.yticks()
plt.grid(True)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

dot product

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z=np.array([1,2,3])
y=np.array([4,5,6])
dot_product=np.dot(z,y)
print(dot_product)
```

Cross product

```python
z=np.array([1,2,3])
y=np.array([4,5,6])
cross_product=np.cross(z,y)
print(cross_product)
```

projection of a vector on v

```python
z=np.array([2,5,56])
v=np.array([8,-6,98])
#magnitude of v vector
magnitut_v =np.sqrt(sum(v**2))
proj_a_on_v=(np.dot(z,v)/magnitut_v**2)*v
print(proj_a_on_v)
```
